```python
x = data.iloc[:,1:-1]
y = data.iloc[:,-1]
```

```python
model = tf.keras.Sequential(
    [tf.keras.layers.Dense(10,input_shape=(3,),activation='relu'),
     tf.keras.layers.Dense(1)]
)
```

```python
model.summary()
```

```python
model.compile(
    optimizer = 'adam',
    loss = 'mse'
    )
```

```python
model.fit(x,y,epochs = 100)
```

```python
test = data.iloc[:10,1:-1]
```

```python
model.predict(test)
```
