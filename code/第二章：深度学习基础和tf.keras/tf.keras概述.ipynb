```python
data = pd.read_csv('../../data/Income1.csv')
```

```python
plt.scatter(data.Education,data.Income)
```

 单变量线性回归：$ f(x) = ax + b $ 表示输入特征与输出值的关系
 预测目标与损失函数

```python
x = data.Education
y = data.Income
```

```python
model = tf.keras.Sequential()
```

```python
model.add(tf.keras.layers.Dense(1,input_shape = (1,))) # ax + b
```

```python
model.summary() 
```

(none,1) none代表的样本的个数

```python
model.compile(
    optimizer='adam',
    loss = 'mse'
)
```

```python
history = model.fit(x,y,epochs = 5000)
```

```python
model.predict(x)
```
